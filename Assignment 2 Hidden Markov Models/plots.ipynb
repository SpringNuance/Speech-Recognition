```python
import matplotlib.pyplot as plt

# Data from the logs
training_passes = list(range(13))  # 0-12
log_likelihood_values = [
    -77.732544,  # Initial hmm-2
    -77.725691, -77.652781, -77.589995,  # hmm-3
    -77.579887, -77.499374, -77.453358,  # hmm-4
    -77.450794, -77.374001, -77.335785,  # hmm-5
    -77.341865, -77.266691, -77.231442   # hmm-6
]

# Plotting the line graph
plt.figure(figsize=(10, 5))
plt.plot(training_passes, log_likelihood_values, marker='o', linestyle='-')
plt.title('HMM Training pass vs Average Log Likelihood')
plt.xlabel('HMM Training pass')
plt.ylabel('Log Likelihood Value')
plt.grid(True)
plt.show()
```
